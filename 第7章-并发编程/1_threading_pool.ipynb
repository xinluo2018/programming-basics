### 线程池使用（扩展学习部分）

```python
import time
import numpy as np
import threading 
from glob import glob
from concurrent.futures import ThreadPoolExecutor
```

```python
### 数据读取函数
def load_data(paths, i_thread):
    print(f"线程 {i_thread} 开始读取数据")
    data = []
    for path in paths:
        arr = np.load(path)  # 读取数据
        arr = arr+1
        data.append(arr)
    print(f"线程 {i_thread} 读取数据完成")
    return len(data)
```

```python
paths = glob('第7章-并发编程/data/*.npy')  # 获取所有数据文件路径
len(paths)
```

#### 使用线程池的两个优点：
1. 可以简单直接获取任务返回结果。 
2. 简化代码（无线程启用.start()，等待线程结束.join()代码）

```python
paths_list = [paths[i::4] for i in range(4)]   # 将数据等分为4份
with ThreadPoolExecutor(max_workers=4) as executor:
    futures = [] ## future为某个任务的执行对象
    ## 提交任务    
    for i in range(4):
        future = executor.submit(load_data, paths_list[i], i)
        futures.append(future)
    ## 获取结果
    results = []
    for future in futures:
        result = future.result()
        results.append(result)
print(f"不同线程读取数据文件数: {results}") 
```
