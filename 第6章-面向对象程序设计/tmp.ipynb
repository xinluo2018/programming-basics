```python
class Student:
    type_student = '大学生'
    def __init__(self, name, age, weight, height):
        self.age = age
        self.weight = weight 
        self.height = height 
    def scores_mean(self, scores):
        print(self.__name)
        score_mean = sum(scores)/len(scores)
        return score_mean
    ## 类方法
    @classmethod
    def print_course(cls, course_list):
        print(course_list)

    ## 静态方法
    @staticmethod
    def compute_bmi(weight, height):
        bmi=weight/height**2
        return bmi

    ## 属性方法
    @property
    def get_bmi(self):
        bmi=self.weight/self.height**2
        return bmi

zhang3 = Student(name='zhang3', age='18', height=1.8, weight=70)
zhang3.scores_mean([78,96,89])
# zhang3.__name 
```
