### 条件与循环

#### 条件

```python
if 1 == 2:
    print('代码运行')
```

```python
num = 10
if num % 2 == 0:
    print("偶数") ## 注意缩进
else:
    print("奇数")

## 简化写法
print("偶数") if num % 2 == 0 else print("奇数")
```

```python
## 多条件判断：elif
num = int(input("请输入一个整数："))
if num > 0:
    print("正数")
elif num < 0: 
    print("负数")
else:
    print("零")
```

```python
## 条件嵌套
num = 10
if num > 0:
    print("正数")
    if num % 2 == 0:
        print("正偶数")
    else:
        print("正奇数")

# ## 逻辑运算符: and, or, not
# if num > 0 and num % 2 == 0:
#     print("正偶数") 
```

```python
## bmi 计算, bmi = 体重(kg) / 身高(m)^2. 
## < 18.5为偏瘦, 18.5~24.9为正常, 25~29.9为偏胖, >=30为肥胖
weight = 100  # kg
height = 1.75  # m
gender = "女"  # 男/女
bmi = weight / (height ** 2)
## 多条件判断+嵌套
if bmi < 18.5:
    if gender == "男":
      print("先生您好，您的bmi指数表明您偏瘦")
    else:
      print("女士您好，您的bmi指数表明您偏瘦")
elif 18.5 <= bmi < 25:
    if gender == "男":
      print("先生您好，您的bmi指数表明您正常")
    else:
      print("女士您好，您的bmi指数表明您正常")
elif 25 <= bmi < 30:  
    if gender == "男":
      print("先生您好，您的bmi指数表明您偏胖")
    else:
      print("女士您好，您的bmi指数表明您偏胖")
else:
    if gender == "男":
      print("先生您好，您的bmi指数表明您肥胖")
    else:
      print("女士您好，您的bmi指数表明您肥胖")

```

#### 循环

##### for...in...循环

```python
for i in range(1, 10): # 生成1到9的数字
    print(i)  # 输出每个数字  

num = 1
for i in range(1,21):  # 左闭右开区间
    num = num+i   # 累加

print('1到20的累加结果是：', num)  # 输出累加结果
```

```python
list1 = [1, 2, 3]  # 列表
for i in list1:
    print(i)  # 输出列表中的每个元素

```

##### while循环

```python
total = 0
while total < 100:  # 当总和小于100时
  num = int(input('请输入一个数字：'))  # 输入数字
  total += num  # 累加输入的数字
  print('当前总和是：', total)  # 输出当前总和
print('总和是：', total)  # 输出总和
```

循环控制

```python
## 循环控制：continue/break 示例
for i in range(10):
    if i == 5: 
        break
    print(i)
```

```python
### 循环嵌套 
for i in range(1, 4):  # 外层循环
    for j in range(1, 4):  # 内层循环
        print(f'外层循环变量i={i}, 内层循环变量j={j}')  # 输出当前的i和j值

## 练习：打印乘法表
for i in range(1, 10):  # 外层循环控制行数
    for j in range(1, i + 1):  # 内层循环控制每行的列数
        print(f'{j} x {i} = {i * j}', end='   ')  # 输出乘法表达式，使用制表符分隔
    print()  # 每完成一行后换行
```

```python
### 可迭代对象：list，string, dict, 支持for循环遍历
num = [10, 20, 30]
for n in num:
    print(n)

greet = "hello"
for char in greet:
    print(char)

student = {'zhang3': 76, 'li4': 80, 'wang5':92}  # 字典
for key in student:
    print(key, ':', student[key])  # 输出字典中的每个键及其对应的值
```

```python
## 列表推导式：生成平方数列表 
squares = [x**2 for x in range(1, 11)]  # 生成1到10的平方列表

## 字典推导式：生成平方数字典
squared_dict = {x: x**2 for x in range(1, 11)}  # 生成1到10的平方字典

## 带条件推导式
even_squares = [x**2 for x in range(1, 11) if x % 2 == 0]  # 生成1到10中偶数的平方列表

```
