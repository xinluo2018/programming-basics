```python
### 库的作用，以中位数计算
def median(nums_list): 
    if not nums_list:
        return None
    nums_list.sort()  ## 排序
    n = len(nums_list)
    mid = n // 2
    if n % 2 == 0:
        return (nums_list[mid - 1] + nums_list[mid]) / 2
    else:
        return nums_list[mid]

nums = [1,3,5,7,10,23,27,9,17]
result = median(nums_list=nums)
print(result)
```

```python
import statistics
result = statistics.median(nums)
print(result)  # 输出: 9.0
```
