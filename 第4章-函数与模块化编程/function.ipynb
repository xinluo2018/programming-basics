### 函数

```python
## 位置参数, 用‘/’规定 
def subs(a, b, /):
    return a - b
subs(2,1)
```

```python
## 关键字参数
def subs(a,b):
    return a - b

c = subs(a=1, b=2)
# print(c)

## 默认参数
def subs(a=3,b=1):
    return a - b
subs(a=2, b=3)
```

```python
## 测试可变参数
def add(*args, **kwargs):
    print('位置参数:', args)
    print('关键字参数:', kwargs)
    return sum(args)+sum(kwargs.values())

add(1, 2, 3, 5,6,7,9,10, a=3, b=4,C=9,d=4)
```

```python
## 返回值
def add(x, y):
    add_result = x + y
    return add_result   # 返回一个值
a = add(1, 2)
print(a)
def compute(x, y):
    add_result = x + y
    sub_result = x - y
    return add_result, sub_result # 返回多个值
a = compute(1, 2)
print(a)
a, b = compute(1, 2)
print(a, b)
```

```python
### lambda 函数
def add(x, y):
    return x + y
print(add(1, 2))  # 调用函数
add = lambda x, y: x + y  # 使用lambda表达式定义函数
print(add(1, 2))  # 调用函数
```

#### 递归函数

```python
def fibonacci(n):
    """递归计算斐波那契数列"""
    if n <= 1:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)

fibonacci(n=5)
```

#### 装饰器

```python
def decorator(func):
    """装饰器示例"""
    def wrapper(a, b):
        print(f'输入数字a和b分别为{a}, {b}')
        func(a, b)
    return wrapper

@decorator
def add_a_b(a,b):
    """简单函数示例"""
    c = a + b
    print(f"The result is {c}!")

@decorator
def subtract_a_b(a,b):
    """简单函数示例"""
    c = a - b
    print(f"The result is {c}!")

add_a_b(a=3,b=4)
subtract_a_b(a=5,b=2)

```

#### 生成器

```python
def my_generator():
    """生成器示例"""
    yield 1
    yield 2
    yield 3

gen = my_generator()
gen2 = (x * x for x in range(5))
for value in gen2:
    print(value)
```

#### 迭代器

```python
list1 = [1, 2, 3, 4, 5]
tuple1 = (1, 2, 3, 4, 5)
list1_iter = iter(list1)
tuple1_iter = iter(tuple1)
next(list1_iter)  # list1_iter是一个迭代器
next(tuple1_iter)  # tuple1_iter是一个迭代器
```
